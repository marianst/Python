Correlação a mapa de calor

```python
dados.info()
```

```python
top_5 = dados[dados.ranking <= 5]
```

```python
top_5[['danceability', 'energy','key','mode','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration' ]]
```

```python
top_5_num = top_5[['danceability', 'energy','key','mode','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration' ]]
```

```python
top_5_num.corr()
```

```python
top_5_corr = top_5_num.corr()
```

https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
heatmap1 = sns.heatmap(top_5_corr, annot=True)
heatmap1.figure.set_size_inches(20, 8)
```

```python
heatmap2 = sns.heatmap(top_5_corr, annot=True,cmap="crest")
heatmap2.figure.set_size_inches(20, 8)
```

```python
heatmap3 = sns.heatmap(top_5_corr, annot=True,cmap=sns.cubehelix_palette(as_cmap=True))
heatmap3.figure.set_size_inches(20, 8)
```
