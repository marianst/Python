##Agrupamento

```python
selecao2_agrupamento = dados[selecao2][['ranking', 'nome_da_musica', 'nome_dos_artistas', 'artista_individual',
       'genero', 'collab', 'data_de_lancamento', 'gravadora','streams','danceability', 'energy']]
```

```python
selecao2_agrupamento.groupby('nome_da_musica').sum('streams').sort_values('streams', ascending=False)
```

```python
selecao2_agrupamento.groupby(['nome_da_musica','artista_individual']).sum('streams').sort_values('streams', ascending = False)
```
