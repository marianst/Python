## Extraindo consultas de um banco de dados.

```python
import sqlite3
import pandas as pd
```

```python
conec_banco = sqlite3.connect('/content/drive/MyDrive/Colab Notebooks/loja_de_roupas_do_joao.db')
```

```python
dados7 = pd.read_sql_query('Select * from produtos;', conec_banco)
```

```python
dados7.head()
```

```python
dados7 = pd.read_sql_query('Select produto,categoria from produtos;', conec_banco)
```

```python
dados7.head()
```

```python
dados7 = pd.read_sql_query('''Select produto
                                    ,categoria
                                from produtos;''', conec_banco)
```

```python
dados7.head()
```

```python
dados7 = pd.read_sql_query('''Select produto
                                    ,categoria
                                from produtos
                                where produto = 'Camiseta';''', conec_banco)
```

```python
dados7.head()
```

```python
dados7 = pd.read_sql_query('Select produto,categoria from produtos where produto = "Camiseta";', conec_banco)
```
